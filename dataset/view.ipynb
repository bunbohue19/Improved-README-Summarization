```python
import pandas as pd
import re

train_df = pd.read_csv('./train.csv', usecols=['readme', 'description'])

max_idx = 0
max = len(train_df['description'][0])
for i, sample in enumerate(train_df['description']):
    if len(sample) > max:
        max = len(sample)
        max_idx = i
        
print(max_idx, max)
```

```python
def preprocessing_description(description):
    if description.endswith('.'):
        description = description[:-1]
    description = re.sub(r"\. ", ", ", description)
    return description + '.'
```

```python
lens = []
for sample in train_df['description']:
    lens.append(len(sample))
print(sum(lens) / len(lens))
```
